#### Implementing Streaming

```python
import anthropic
from dotenv import load_dotenv

load_dotenv()

client = anthropic.Anthropic()

model = "claude-sonnet-4-5-20250929"

stream = client.messages.create(
    model=model,
    max_tokens=1000,
    stream=True,
    messages = [
    {
        "role": "user",
        "content": "What is Pluralsight? What is Pluralsight? Answer in a single, short, sentence."
    }
]  
)

for event in stream:
    print(event)
```

#### Simplified Streaming

```python
import anthropic
from dotenv import load_dotenv

load_dotenv()
client = anthropic.Anthropic()
model = "claude-sonnet-4-5-20250929"

messages = [
    {
        "role": "user",
        "content": "What is Pluralsight?"
    }
]

with client.messages.stream(
    model=model,
    max_tokens=1000,
    messages=messages
) as stream:
    for text in stream.text_stream:
        print(text, end="")
```
