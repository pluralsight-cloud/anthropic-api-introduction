#### Send a Message

```python
import anthropic
from dotenv import load_dotenv

load_dotenv()

client = anthropic.Anthropic()
model = "claude-sonnet-4-5-20250929"

message = client.messages.create(
    model= model,
    max_tokens=1000,
    temperature=1,
    messages=[
        {
            "role": "user",
            "content": "What is Pluralsight? Answer in a single, short, sentence."
        }
    ]
)
print(message.content[0].text)
```

#### Send an Additional Message

```python
message = client.messages.create(
    model="claude-sonnet-4-5-20250929",
    max_tokens=1000,
    temperature=1,
    messages=[
        {
            "role": "user",
            "content": "In a single sentence, tell me more."
        }
    ]
)
print(message.content[0].text)
```

#### Conversations

```python
def add_message(messages, role, text):
    message = {"role": role, "content": text}
    messages.append(message)

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    return message.content[0].text

# Create a list of to hold the messages
messages = []

# Add the first message from the user
add_message(messages, "user", "What is Pluralsight? Answer in a single, short, sentence.")

# Get a response from Claude
response = chat(messages)

# Add Claude's response to list of messages
add_message(messages, "assistant", response)
```

#### View the Messages

```python
print(messages)
```

#### Add a Follow-up Message

```python
# Add a follow-up question
add_message(messages, "user", "In a single sentence, tell me more.")

# Get the follow-up response with full conversation history
additional_response = chat(messages)

# Add Claude's latest response to list of messages
add_message(messages, "assistant", additional_response)

# Print the last message
print(messages[-1]['content'])
```
